```python
train = pd.read_csv('/kaggle/input/dacon1/train.csv')
train
```

```python
test = pd.read_csv('/kaggle/input/dacon1/test.csv')
test
```

```python
train['class'].value_counts()
```

```python
alldata = pd.concat([train,test])
alldata
```

```python
alldata2 = alldata.drop(columns=['class','id'])
alldata2
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
snp_list = alldata2.columns[4:]
for i in snp_list:
    alldata2[i] = le.fit_transform(alldata2[i])
alldata2
```

```python
train2 = alldata2[0:len(train)]
test2 = alldata2[len(train):]
```

```python
from catboost import CatBoostClassifier
cbc = CatBoostClassifier(verbose = 100)
cbc.fit(train2,train['class'])
result = cbc.predict(test2)
```

```python
sub = pd.read_csv('/kaggle/input/dacon1/sample_submission.csv')
sub['class'] = result
sub['class'].value_counts()
```

```python
sub.to_csv('sub.csv', index = 0)
```
